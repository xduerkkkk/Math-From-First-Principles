```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. 定义绘图函数 ---
# 我们把绘图的代码封装成一个函数，方便以后重复使用
def plot_vectors(vectors, colors):
    """
    绘制一组向量。
    vectors: 向量列表，例如 [[1,0], [0,1]]
    colors: 颜色列表，例如 ['r', 'b']
    """
    for i in range(len(vectors)):
        # 从原点 (0,0) 开始绘制箭头
        plt.quiver(0, 0, vectors[i][0], vectors[i][1], angles='xy', scale_units='xy', scale=1, color=colors[i])

# --- 2. 定义我们的原始空间 ---
# 标准基向量
i_hat = np.array([1, 0])
j_hat = np.array([0, 1])

# 原始基向量列表和颜色
original_basis = [i_hat, j_hat]
colors = ['red', 'blue'] # i_hat 是红色, j_hat 是蓝色

# --- 3. 绘图 ---
plt.figure(figsize=(6, 6)) # 创建一个 6x6 英寸的画布

# 绘制基向量
plot_vectors(original_basis, colors)

# 设置坐标轴范围，让图像居中好看
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# 画网格线
plt.grid()

# 保持坐标轴比例为1:1
plt.gca().set_aspect('equal', adjustable='box')

# 显示图像
plt.title("Original Space")
plt.show()
```

```python
A = np.array([
    [1, -1],
    [0.5, 2]
])

# --- 5. 应用变换 ---
# 用矩阵 A 去乘以我们的基向量
# NumPy 中，矩阵乘法可以用 @ 符号 (或者 np.dot() 函数)
i_hat_transformed = A @ i_hat
j_hat_transformed = A @ j_hat

# 变换后的基向量列表
transformed_basis = [i_hat_transformed, j_hat_transformed]

# --- 6. 再次绘图，对比变换前后 ---
plt.figure(figsize=(12, 6)) # 创建一个更大的画布，12x6 英寸

# 左边的子图：原始空间
plt.subplot(1, 2, 1) # (1行, 2列, 第1个图)
plot_vectors(original_basis, colors)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Original Space")

# 右边的子图：变换后的空间
plt.subplot(1, 2, 2) # (1行, 2列, 第2个图)
plot_vectors(transformed_basis, colors)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Transformed Space")

plt.show()

# 我们也可以打印出变换后的基向量坐标，看看它们具体是多少
print("Original i_hat:", i_hat)
print("Transformed i_hat:", i_hat_transformed)
print("Original j_hat:", j_hat)
print("Transformed j_hat:", j_hat_transformed)
```

```python
# --- 7. 定义一个任意向量 v ---
v = np.array([1, 2]) # 让我们看看向量 (1,2) 会有什么遭遇

# --- 8. 计算 v 的变换 ---
# 变换前的 v, 可以看作是基向量的线性组合
# v = 1 * i_hat + 2 * j_hat
v_transformed = A @ v

# 也可以通过变换后的基向量来计算，结果是一样的
# v_transformed_manual = 1 * i_hat_transformed + 2 * j_hat_transformed
# print("Manually calculated transformed v:", v_transformed_manual) # 你可以取消这行的注释来验证

# --- 9. 在一张图上，同时画出变换前后的所有东西 ---
plt.figure(figsize=(8, 8))

# 绘制变换前的基向量
plot_vectors(original_basis, ['#FFDDDD', '#DDDDFF']) # 用浅色表示原始基向量

# 绘制变换后的基向量
plot_vectors(transformed_basis, ['red', 'blue']) # 用深色表示变换后基向量

# 绘制变换前的向量 v
plot_vectors([v], ['#DDFFDD']) # 用浅绿色表示原始 v

# 绘制变换后的向量 v_transformed
plot_vectors([v_transformed], ['green']) # 用深绿色表示变换后 v

# 设置坐标轴范围
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Vector Transformation")
plt.show()

print("\nOriginal vector v:", v)
print("Transformed vector v:", v_transformed)
```
