```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return (x**2 - 4) / (x - 2)

x = np.linspace(0, 4, 400)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = (x^2-4)/(x-2)')
# 在 x=2 的位置画一个空心圆
plt.plot(2, 4, 'o', markerfacecolor='white', markeredgecolor='red', markersize=8, label='未定义的点： x=2')
plt.title('A Function with a "Hole"')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# --- 字体设置 ---
# 已经确认 simhei.ttf 是有效的
font_path = 'C:/Windows/Fonts/simhei.ttf'
my_font = fm.FontProperties(fname=font_path)

# --- 统一风格 ---
plt.style.use('seaborn-v0_8-whitegrid')

# ==========================================================
# --- 图1: 完美情况 (极限 = 函数值) ---
# ==========================================================
fig1, ax1 = plt.subplots(figsize=(8, 6))

x1 = np.linspace(0, 4, 100)
y1 = x1**2

ax1.plot(x1, y1, label=r'$f(x) = x^2$')
ax1.plot(2, 4, 'o', markersize=10, color='blue', label=r'事实(函数值): f(2) = 4')

# 【修正】在这里应用你的中文字体
ax1.set_title('图1: 完美情况 (极限 = 函数值)', fontproperties=my_font, fontsize=16)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
# 【修正】legend 使用 'prop' 参数来设置字体
ax1.legend(prop=my_font, fontsize=12)
# 【修正】text 也需要设置字体
ax1.text(2.1, 3.5, r'趋势: $\lim_{x \to 2} x^2 = 4$', fontproperties=my_font, fontsize=14, color='darkgreen')

plt.show()

# ==========================================================
# --- 图2: 意外情况 (极限 ≠ 函数值) ---
# ==========================================================
fig2, ax2 = plt.subplots(figsize=(8, 6))

x2 = np.linspace(0, 4, 100)
y2 = x2

ax2.plot(x2, y2, label='函数趋势线 y=x')
ax2.plot(2, 2, 'o', markersize=10, markerfacecolor='white', markeredgecolor='blue', label=r'趋势(极限): $\lim_{x \to 2} f(x) = 2$')
ax2.plot(2, 3.5, 'o', markersize=10, color='red', label=r'事实(函数值): $f(2)=3.5$')

# 【修正】在这里应用你的中文字体
ax2.set_title('图2: 意外情况 (极限 ≠ 函数值)', fontproperties=my_font, fontsize=16)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('f(x)', fontsize=12)
# 【修正】legend 使用 'prop' 参数来设置字体
ax2.legend(prop=my_font, fontsize=12)

plt.show()

# ==========================================================
# --- 图3: 空洞情况 (极限存在, 函数值不存在) ---
# ==========================================================
fig3, ax3 = plt.subplots(figsize=(8, 6))

x3 = np.linspace(0, 4, 100)
y3 = (x3**2 - 4) / (x3 - 2)

ax3.plot(x3, y3, label=r'$f(x) = \frac{x^2-4}{x-2}$')
ax3.plot(2, 4, 'o', markersize=10, markerfacecolor='white', markeredgecolor='red', label='事实: x=2 处无定义')

# 【修正】同样，为保持统一，给图3也明确加上字体设置
ax3.set_title('图3: 空洞情况 (极限存在, 函数值不存在)', fontproperties=my_font, fontsize=16)
ax3.set_xlabel('x', fontsize=12)
ax3.set_ylabel('f(x)', fontsize=12)
# 【修正】legend 使用 'prop' 参数来设置字体
ax3.legend(prop=my_font, fontsize=12)
# 【修正】text 也需要设置字体
ax3.text(2.1, 3.5, r'趋势: $\lim_{x \to 2} f(x) = 4$', fontproperties=my_font, fontsize=14, color='darkgreen')

plt.show()
```
